# Instantiating corpora and corpus class

```python
# import ekorpkit.ekonf as eKonf
from ekorpkit import eKonf
from pprint import pprint
```

## Instantiating corpora with an ekorpkit config

```python
cfg = eKonf.compose(config_group='corpus')
print('Config type:', type(cfg))

cfg['name'] = ['bok_minutes', 'esg_report']
cfg['data_dir'] = '/workspace/data/datasets/corpus/ekorpkit'
pprint(cfg)
crps = eKonf.instantiate(cfg)
```

```python
crps['bok_minutes'].data
```

```python
print(crps)
```

```python
crps.concat_corpora()
```

## Instantiating a corpus with an ekorpkit config

```python
cfg = eKonf.compose(config_group='corpus=corpus')
print('Config type:', type(cfg))

cfg['name'] = 'bok_minutes'
cfg['data_dir'] = '/workspace/data/datasets/corpus/ekorpkit/bok_minutes'
pprint(cfg)
crps = eKonf.instantiate(cfg)
```
