## Process a pipeline  with the ekorpkit configs

```python
cfg = eKonf.compose(config_group='pipeline=dataframe')
cfg.verbose = True
cfg.data_dir = './data/fomc_minutes'
cfg.data_file = 'fomc_minutes.csv'
cfg._pipeline_ = ['load_dataframe', 'summary_stats']
cfg.summary_stats.output_file = 'fomc_minutes_stats.yaml'
df = eKonf.instantiate(cfg)
```
